```julia
data = CSV.read("data.csv")
```

```julia
keep = findall(!ismissing, data[:age])
r_df = data[keep, :]
keep = findall(!ismissing, r_df[:grade])
r_df = r_df[keep, :]
keep = findall(!ismissing, r_df[:sex])
r_df = r_df[keep, :]
keep = findall(!ismissing, r_df[:q41])
data1 = r_df[keep, :]
```

```julia
data2 = data1
for i in names(data2)[11:length(names(data2))] 
    data2[i] = passmissing(convert).(Int8,data2[i])
end
for i in names(data2) 
    println(eltype(data2[i]))
end
```

```julia
data2 = select!(data2, Not(:Column1))
data2 = select!(data2, Not(:year))
```

```julia
for i in names(data2)
    println(i, " which is ", eltype(data2[i]))
end 
```

```julia
for i in names(data2)[11:length(names(data2))] 
    data2[i] = passmissing(convert).(Integer,data2[i])
end
```

```julia
data2[:grade] = passmissing(convert).(Integer,data2[:grade])
data2 = select!(data2, Not(:bmipct))
data2[:qnobese] = passmissing(convert).(Integer,data2[:qnobese])
```

```julia
data3 = data2
for i in names(data3)
    println(i)
    if (length(unique(data3[i]))==3)
        data3[i] = passmissing(convert).(Bool,(data3[i].-=1))
    end
end
data3["qn41"] = passmissing(convert).(Bool,(data3["qn41"].-=1))
```
